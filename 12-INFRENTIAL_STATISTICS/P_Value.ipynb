## P-Value and Significance Level

### What is a P-Value?

The **P-value (probability value)** is the probability of obtaining results **at least as extreme as the observed results**, assuming that the **null hypothesis (H₀)** is true.

It helps determine whether the observed data provides enough evidence to **reject the null hypothesis**.

---

### Interpretation of P-Value

- A **small P-value** (typically ≤ 0.05) indicates **strong evidence against H₀**, so you **reject the null hypothesis**.  
- A **large P-value** (> 0.05) indicates **weak evidence against H₀**, so you **fail to reject H₀**.

| P-Value Range | Decision | Interpretation |
|----------------|-----------|----------------|
| P ≤ 0.01 | Reject H₀ | Very strong evidence against H₀ |
| 0.01 < P ≤ 0.05 | Reject H₀ | Moderate evidence against H₀ |
| 0.05 < P ≤ 0.10 | Fail to reject H₀ | Weak evidence against H₀ |
| P > 0.10 | Fail to reject H₀ | No evidence against H₀ |

---

### Significance Level (α)

The **significance level (α)** is a threshold set by the researcher to decide whether to reject H₀.  
It represents the **maximum probability of making a Type I error** (rejecting a true null hypothesis).

Common significance levels are:

| Significance Level (α) | Confidence Level | Meaning |
|-------------------------|------------------|----------|
| 0.10 | 90% | Tolerates 10% chance of error |
| 0.05 | 95% | Most commonly used in research |
| 0.01 | 99% | Very strict; less chance of error |

---

### Example

> Suppose you test whether a new teaching method improves student performance.  
> The test gives a **P-value of 0.03** at a **significance level (α) of 0.05**.

## Example: Testing Whether a Coin is Fair

---

### Null Hypothesis (H₀)
The coin is **fair**.  
\[
H₀ : \text{The coin is fair.}
\]

---

### Alternate Hypothesis (H₁)
The coin is **not fair**.  
\[
H₁ : \text{The coin is not fair.}
\]

---

**Decision:**  
Since P = 0.03 < 0.05, we **reject H₀**.

**Conclusion:**  
There is **significant evidence** that the new teaching method improves student performance.

### Experiment
The coin is tossed **100 times** and the number of heads is recorded.

A normal distribution curve is assumed, where:

- The **center region** represents that the **coin is fair**.  
- The **extreme ends (tails)** on both sides represent the **rejection regions**, where the coin is **not fair**.  

Each rejection area covers **0.025** probability (total α = 0.05).

---

### Significance Value (α)
\[
\alpha = 0.05
\]
This means there is a **5% chance of rejecting a true null hypothesis** (Type I error).

The **Confidence Interval (CI)** is:
\[
CI = 1 - \alpha = 1 - 0.05 = 0.95
\]
Hence, we are **95% confident** about our decision.

---

### Conclusion

If  
\[
p < \alpha
\]
then **Reject the Null Hypothesis (H₀)**  
→ The coin is **not fair**.

Else  
**Fail to reject H₀**  
→ The coin is **fair**.

---

### Summary Table

| Condition | Decision | Interpretation |
|------------|-----------|----------------|
| p < 0.05 | Reject H₀ | Coin is **not fair** |
| p ≥ 0.05 | Fail to reject H₀ | Coin is **fair** |

---

**Visualization Summary:**  
- Middle region (95%) → Coin is fair  
- Extreme regions (each 2.5%) → Coin is not fair (Rejection areas)
