# Probability Distribution Function

A **Probability Distribution Function (PDF)** describes how probabilities are distributed over the values of a **random variable**.  
It shows how likely each outcome is and helps visualize how probabilities are assigned to different possible values.

---

## Types of Probability Distribution Functions

There are **three main types** of probability distribution functions:

1. **Probability Mass Function (PMF)**  
   Used for **discrete random variables** such as dice rolls or coin tosses.  
   \( P(X = x) \) = Probability that random variable \( X \) takes the value \( x \).

2. **Probability Density Function (PDF)**  
   Used for **continuous random variables** such as height, weight, or age.  
   The **area under the curve** gives the probability that the variable falls within that interval.

3. **Cumulative Distribution Function (CDF)**  
   Represents the **cumulative probability** up to a certain value \( x \).  
   \( F(x) = P(X \leq x) \)

---

## Example 1: Rolling a Dice (Discrete Variable)

**Random Variable:** \( X = \) outcome of rolling one fair dice  
Possible values: \( 1, 2, 3, 4, 5, 6 \)

### PMF Calculation
Each face has equal probability:

\[
P(X = x) = \frac{1}{6}, \quad \text{for } x = 1, 2, 3, 4, 5, 6
\]

| X | 1 | 2 | 3 | 4 | 5 | 6 |
|---|---|---|---|---|---|---|
| P(X = x) | 1/6 | 1/6 | 1/6 | 1/6 | 1/6 | 1/6 |



### CDF Calculation
Cumulative probability up to a given value \( x \):

\[
F(x) = P(X \leq x) = \sum_{k=1}^{x} P(X = k)
\]

Example:  
\[
P(X \leq 3) = P(1) + P(2) + P(3) = \frac{3}{6} = 0.5
\]

## Example 2: Random Ages (Continuous Variable)

**Random Variable:** \( X = \) Age of a randomly selected person  
Assume a **normal distribution** with  
Mean \( \mu = 40 \) and Standard Deviation \( \sigma = 7 \).

### PDF Calculation

The formula for the probability density function is:

\[
f(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{(x-\mu)^2}{2\sigma^2}}
\]

Example:
\[
f(40) = \frac{1}{7\sqrt{2\pi}} e^{-\frac{(40-40)^2}{2(7)^2}} \approx 0.057
\]

### CDF Calculation

Cumulative probability up to a certain age:

\[
F(x) = P(X \leq x)
\]

Example:  
\[
P(X \leq 35) = 0.31, \quad P(X \leq 40) = 0.50, \quad P(X \leq 45) = 0.69
\]



Interpretation:
- 31% of people are aged ≤ 35  
- 50% are aged ≤ 40 (median)  
- 69% are aged ≤ 45

## Common Probability Distributions

### Discrete:
- Bernoulli Distribution  
- Binomial Distribution  
- Poisson Distribution   

### Continuous:
- Uniform Distribution  
- Normal (Gaussian) Distribution  
- Exponential Distribution  
- Log-Normal Distribution  
- Pareto Distribution  

---

### Summary

| Type | Variable | Function | Key Concept |
|------|-----------|-----------|--------------|
| PMF | Discrete | \( P(X=x) \) | Exact probability of a value |
| PDF | Continuous | \( f(x) \) | Density; area = probability |
| CDF | Both | \( F(x)=P(X≤x) \) | Cumulative probability up to x |

---

**In short:**  
- **PMF:** Exact probability (discrete)  
- **PDF:** Probability density (continuous)  
- **CDF:** Cumulative probability (both)  
- **Area under PDF = Probability**
