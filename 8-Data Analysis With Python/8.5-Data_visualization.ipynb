# DATA VISUALIZATION 
Matplotlib is a widely-used Python library used for creating static, animated and interactive data visualizations. It is built on the top of NumPy and it can easily handles large datasets for creating various types of plots such as line charts, bar charts, scatter plots, etc.

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [1,4,9,16,25]

# create a line plot 
plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title("Basic Line Plot")
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [1,4,9,16,25]

plt.plot(x, y, color='red', linestyle='--', marker='o', linewidth=2, markersize=9)
plt.grid(True)
```

```python
# multiple plots 
# sample data 

x = [1,2,3,4,5]
y1 = [1,4,9,16,25]
y2 = [1,2,3,4,5]

plt.figure(figsize=(9,7))

plt.subplot(2,2,1) # one rwo two coloumns and thrid parameter is position of the plt whic position we are ploting that.
plt.plot(x,y1,color='green')
plt.title("Plot 1")

plt.subplot(2,2,2) 
plt.plot(x,y2,color='pink')
plt.title("Plot 2")

plt.subplot(2,2,3) 
plt.plot(x,y1,color='red')
plt.title("Plot 3")

plt.subplot(2,2,4) 
plt.plot(x,y2,color='blue')
plt.title("Plot 4")
```

# BAR PLOT

```python
categories = ['A','B','C','D']
values = [5,7,6,3]

# create a bar plot 
plt.bar(categories,values,color='red')
plt.xlabel('categories')
plt.ylabel('Values')
plt.title('Bar Plot')
plt.show()
```

# HISTOGRAM

plt.hist() divides your continuous data into bins (intervals) and counts how many data points fall in each bin.

```python

ages = [10, 12, 15, 20, 22, 25, 27, 30, 35, 36, 38, 40, 41, 42, 45]

plt.hist(ages, bins=5, color='skyblue', edgecolor='black')
plt.xlabel("Age range")
plt.ylabel("Number of people")
plt.title("Age Distribution Histogram")
plt.show()
```

# Scatter Plot
A scatter plot is used to show the relationship or correlation between two continuous variables.
Each point on the graph represents one observation — its position depends on its x and y values.

```python

# Example: Study hours vs Exam scores
x = [1, 2, 3, 4, 5, 6, 7, 8]       # Study hours
y = [35, 42, 50, 58, 65, 70, 75, 80]  # Exam marks

plt.scatter(x, y, color='orange', marker='o', s=100, edgecolor='black')

plt.xlabel("Study Hours")
plt.ylabel("Exam Marks")
plt.title("Study Hours vs Marks (Scatter Plot)")
plt.grid(True)
plt.show()
```

# pie chart 
A pie chart shows how different parts contribute to a whole.
It divides data into slices, where each slice represents a percentage of the total.

In short:

It helps you visualize proportions or composition of categories.

```python
# Data
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

# Highlight the first slice (Python)
explode = [0.1, 0, 0, 0]

plt.pie(sizes, labels=labels, explode=explode,
        autopct='%1.1f%%', shadow=True, startangle=90)

plt.title("Programming Language Popularity")
plt.show()
```

```python
import pandas as pd
sales_data_df= pd.read_csv('data_1.csv')
print(sales_data_df)
```

```python
sales_data_df.info()
```

```python
# plot total sales by product 
total_sales_by_product = sales_data_df.groupby('Product')['Sales'].sum()
print(total_sales_by_product)
```

```python
total_sales_by_product.plot(kind='bar',color='orange')
```

```python
# ploting sales tren over time 

sales_trend = sales_data_df.groupby('Date')['Sales'].sum().reset_index()
plt.plot(sales_trend['Date'],sales_trend['Sales'])
```

```python

import matplotlib.dates as mdates


# Ensure Date column is in datetime format
sales_data_df['Date'] = pd.to_datetime(sales_data_df['Date'])

# Group by date and sum sales
sales_trend = sales_data_df.groupby('Date')['Sales'].sum().reset_index()

# Plot
plt.figure(figsize=(12,6))
plt.plot(sales_trend['Date'], sales_trend['Sales'], color='royalblue', marker='o', linewidth=2, markersize=5)

# Axis labels and title
plt.title('Sales Trend Over Time', fontsize=14, fontweight='bold')
plt.xlabel('Date', fontsize=12)
plt.ylabel('Total Sales', fontsize=12)

# Format x-axis to show only monthly ticks
plt.gca().xaxis.set_major_locator(mdates.MonthLocator(interval=1))   # show one tick per month
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))  # format as 'Jan 2023'

# Rotate date labels for better visibility
plt.xticks(rotation=45)

# Add grid and layout
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()

# Show plot
plt.show()
```
