
<h1 align="center">STATICTICS</h1>

## Definition

Statistics is the science of collecting, analyzing, and interpreting data to uncover patterns and make decisions.  
In data science, it acts as the backbone for understanding data and building reliable models.

**Statistics helps to:**
- Summarize data using measures like mean, median, and variance  
- Model uncertainty with probability and distributions  
- Test hypotheses (e.g., A/B testing)  
- Find relationships through regression and correlation  

---

## Data

**Definition:**  
Data refers to raw facts, figures, or information collected from various sources, which by themselves may not have meaning.

When this data is processed, cleaned, and analyzed, it helps to uncover patterns, insights, and support decision-making.

---

## Applications of Statistics

1. **Data Exploration and Summarization**  
   - Involves collecting, organizing, and summarizing data.  
   - Helps understand trends, patterns, and relationships.  
   - Example: Using mean, median, mode, and visualizations like histograms or box plots.

2. **Model Building and Validation**  
   - Statistics helps create predictive or explanatory models.  
   - Example: Regression models, classification models, etc.

3. **Statistical Analysis**  
   - Analyzing sample data to make conclusions about the population.  
   - Example: Confidence intervals, hypothesis testing.

4. **Hypothesis Testing**  
   - Used to test assumptions or claims about data.  
   - Helps in decision-making using probability and statistical evidence.

5. **Optimization and Efficiency**  
   - Used to optimize processes and improve performance.  
   - Example: Reducing manufacturing defects, maximizing profit, or minimizing costs.

6. **Reporting**  
   - Presenting analyzed data clearly through reports, dashboards, and visualizations.

## Types of Statistics

### 1. Descriptive Statistics

**Definition:** Organizing, summarizing, and presenting data in a meaningful way.

**Main Components:**
- **Measures of Central Tendency:** Mean, Median, Mode  
- **Measures of Dispersion:** Variance, Standard Deviation  
- **Data Distribution:** Histogram, Box Plot, Pie Chart, PDF, PMF  
- **Summary Statistics:** Minimum, Q1, Median, Q3, Maximum  

### 2. Inferential Statistics

**Definition:** Making predictions or generalizations about a population using sample data.

**Key Concepts:** Sample → Population (Inference)

**Main Components:**
- Hypothesis Testing  
- P-value  
- Confidence Interval  
- Statistical Tests: Z-test, t-test, ANOVA, Chi-square  

---

## Population and Sample Data

### Population
**Definition:** Entire set of individuals or objects of interest in a particular study.

**Characteristics:**
- Complete set (contains all observations)  
- Parameter: summarizes entire population (mean, variance)

### Sample
**Definition:** Subset of population used to represent the whole.

**Characteristics:**
- Represents a portion of population  
- Statistic: summarizes sample data (sample mean, sample variance)  
- Should be randomly selected to avoid bias  

**Examples:**
- School study: All students = population; 50 students = sample  
- Market research: All consumers = population; 500 = sample  
- Medical study: All patients = population; 200 = sample  

---

## Types of Sampling Techniques

### 1. Probability Sampling  
Every member of population has a known and equal chance of being selected.  
**Examples:** Simple Random, Stratified, Cluster, Systematic Sampling

### 2. Non-Probability Sampling  
Not every member has equal chance of being selected.  
**Examples:** Convenience, Judgmental, Snowball, Quota, Voluntary Sampling

## Types of Data

### Qualitative Data (Categorical)
Describes qualities, categories, or attributes (not numbers).

| Type | Description | Examples |
|------|--------------|-----------|
| **Nominal** | Names or labels without order | Gender, Color, City |
| **Ordinal** | Ordered categories, ranks not measurable | Rating, Education level |

### Quantitative Data (Numerical)
Represents measurable quantities.

| Type | Description | Examples |
|------|--------------|-----------|
| **Discrete** | Whole, countable values | Students, Cars, Goals |
| **Continuous** | Any value in range (decimals allowed) | Height, Weight, Time |

---

## Scales of Measurement

1. **Nominal Scale**  
   - Used for labeling variables without order.  
   - Examples: Gender, Blood Group, Color

2. **Ordinal Scale**  
   - Ordered categories, but differences not measurable.  
   - Examples: Rating, Education level

3. **Interval Scale**  
   - Ordered data with equal intervals but no true zero.  
   - Examples: Temperature (°C), Calendar years

4. **Ratio Scale**  
   - Equal intervals + true zero point.  
   - Examples: Height, Weight, Age, Income, Time
