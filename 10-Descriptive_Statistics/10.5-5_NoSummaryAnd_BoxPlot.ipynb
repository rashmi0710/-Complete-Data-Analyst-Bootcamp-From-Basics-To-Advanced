<h1 align="center">5 NUMBER SUMMARY AND BOX PLOT<h1>

#### **Five-Number Summary**

### Definition
A **five-number summary** is a statistical measure that describes a dataset using five key values:
1. **Minimum** — the smallest value in the dataset  
2. **First Quartile (Q1)** — the 25th percentile  
3. **Median (Q2)** — the 50th percentile or middle value  
4. **Third Quartile (Q3)** — the 75th percentile  
5. **Maximum** — the largest value in the dataset (excluding extreme outliers)

It provides a quick overview of the data’s **central tendency**, **spread**, and **range**, and is often represented visually using a **box plot**.

---

### **Example**

Given dataset:  
`[1, 2, 2, 2, 3, 3, 4, 5, 5, 6, 6, 6, 6, 7, 8, 8, 9, 27]`

Number of observations, **n = 18**

---

### **Step 1: Identify Outliers**
To check outliers, we use the **Interquartile Range (IQR)** method:  
\[
\text{IQR} = Q3 - Q1
\]

From earlier:  
- Q1 = 3  
- Q3 = 7  
\[
\text{IQR} = 7 - 3 = 4
\]

Any data point greater than \( Q3 + 1.5 \times IQR = 7 + 6 = 13 \)  
or less than \( Q1 - 1.5 \times IQR = 3 - 6 = -3 \) is considered an **outlier**.

Hence, **27 > 13 → Outlier**

---

### **Step 2: Calculate Five-Number Summary (Excluding Outlier 27)**

Clean dataset (excluding outlier):  
`[1, 2, 2, 2, 3, 3, 4, 5, 5, 6, 6, 6, 6, 7, 8, 8, 9]`  
**n = 17**

- **Minimum = 1**  
- **Maximum = 9**

\[
Q2 = \text{Median} = 6 \text{ (9th value)}
\]

\[
Q1 = \text{Median of lower half } [1, 2, 2, 2, 3, 3, 4, 5] = \frac{2 + 3}{2} = 2.5
\]

\[
Q3 = \text{Median of upper half } [6, 6, 6, 7, 8, 8, 9, 27] = \frac{7 + 8}{2} = 7.5
\]
(Since we excluded 27, the upper half becomes `[6, 6, 6, 7, 8, 8, 9]` → Q3 = 7)

---

### **Final Five-Number Summary**

| Statistic | Symbol | Value |
|------------|---------|--------|
| Minimum | – | 1 |
| First Quartile | Q1 | 2.5 |
| Median | Q2 | 6 |
| Third Quartile | Q3 | 7 |
| Maximum | – | 9 |

---

### **Interpretation**
- The **middle 50%** of the data lies between **2.5** and **7**.  
- The **range (without outlier)** is \( 9 - 1 = 8 \).  
- **27** is an **outlier**, which lies far outside the normal spread of data.


<h1 align="center">Box Plot</h1>

### **Definition**
A **Box Plot** (also called a **Box-and-Whisker Plot**) is a graphical representation of a dataset based on its **five-number summary**:
- Minimum  
- First Quartile (Q1)  
- Median (Q2)  
- Third Quartile (Q3)  
- Maximum  

It helps to visualize the **spread**, **central tendency**, and **outliers** in the data.

---

### **Example Dataset**
`[1, 2, 2, 2, 3, 3, 4, 5, 5, 6, 6, 6, 6, 7, 8, 8, 9, 27]`

After removing the outlier (**27**), the dataset becomes:  
`[1, 2, 2, 2, 3, 3, 4, 5, 5, 6, 6, 6, 6, 7, 8, 8, 9]`

---

### **Five-Number Summary**

| Statistic | Symbol | Value |
|------------|---------|--------|
| Minimum | – | 1 |
| First Quartile | Q1 | 2.5 |
| Median | Q2 | 6 |
| Third Quartile | Q3 | 7 |
| Maximum | – | 9 |

### **Interpretation**

- The **middle 50%** of data lies between **Q1 (2.5)** and **Q3 (7)**.  
- The **median (6)** divides the dataset into two equal halves.  
- The **range** of the data (without outlier) is \( 9 - 1 = 8 \).  
- The value **27** is an **outlier**, which lies far beyond the upper whisker.

---

### **Key Uses**
- To quickly identify **outliers**  
- To understand the **distribution** and **spread** of data  
- To compare **different datasets** visually
