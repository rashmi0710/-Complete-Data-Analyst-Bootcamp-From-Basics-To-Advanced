<h1 align="center">PERCENTILES AND QUARTILES</h1>

### Definition
A percentage is a way to express a number or ratio as a fraction of 100, meaning *"per hundred"*.

### Example
Given the list:  
`[1, 2, 3, 4, 5, 6]`  

Number of odd numbers = 3  

$
\text{Percentage of odd numbers in this group} = \frac{3}{6} \times 100 = 50\%
$

#### **Percentiles**

### Definition
A **percentile** is a value below which a given percentage of observations in a dataset fall.

### Example
Given the dataset:  
`{2, 2, 3, 4, 5, 5, 6, 7, 8, 8, 8, 9, 9, 10}`  

To find the **percentile**, the data must be arranged in ascending order (which it already is).  
For example:  
- The **50th percentile (P50)** represents the value below which 50% of the data lie — this is also known as the **median**.  
- The **25th percentile (P25)** represents the value below which 25% of the data lie.  
- The **75th percentile (P75)** represents the value below which 75% of the data lie.

---

#### **Quartiles**

### Definition
**Quartiles** divide a dataset into four equal parts, each containing 25% of the observations.

- **Q1 (First Quartile)** → 25th Percentile → 25% of data lies below this value  
- **Q2 (Second Quartile)** → 50th Percentile → This is the **median**  
- **Q3 (Third Quartile)** → 75th Percentile → 75% of data lies below this value  

### Example
For the dataset:  
`{2, 2, 3, 4, 5, 5, 6, 7, 8, 8, 8, 9, 9, 10}`  

Number of observations, **n = 14**

$
Q1 = \text{Value at } \frac{(n+1)}{4} = \text{Value at } \frac{15}{4} = 3.75^{th} \text{ position} \approx 4
$ 

$
Q2 = \text{Value at } \frac{(n+1)}{2} = \text{Value at } 7.5^{th} \text{ position} \approx 6.5
$  

$
Q3 = \text{Value at } \frac{3(n+1)}{4} = \text{Value at } \frac{45}{4} = 11.25^{th} \text{ position} \approx 8.5
$

So,  
- **Q1 = 4**  
- **Q2 = 6.5 (Median)**  
- **Q3 = 8.5**

---

### Relation Between Quartiles and Percentiles
- Q1 = P25  
- Q2 = P50  
- Q3 = P75
